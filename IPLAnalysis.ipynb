```python
#find the batsman who has more runs and give top 5 batsmen
runs.sum()['batsman_runs'].sort_values(ascending=False).head(5) # Grouping by 'batsman', summing 'batsman_runs', and displaying the top 5 batsmen with the most runs
```

```python
#find the batsman who has more 4s and give top 5 batsmen
mask=delivery['batsman_runs']==4  # Creating a mask for rows where 'batsman_runs' equals 4
new_delivery=delivery[mask] # Filtering the delivery DataFrame to include only rows where the mask condition is met
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(5)  # Grouping by 'batsman', counting 'batsman_runs', and displaying the top 5 batsmen with the most 4s   
```

```python
mask_vk=delivery['batsman']=='V Kohli' # Creating a mask for rows where the 'batsman' is 'V Kohli'
delivery[mask_vk].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head() # Grouping by 'bowling_team' and summing 'batsman_runs' for 'V Kohli', then sorting in descending order
```

```python
def run_scored(batsman):
    mask=delivery['batsman']==batsman # Creating a mask for rows where the 'batsman' is the specified batsman
    return delivery[mask].groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)# Calling the function to get runs scored by 'V Kohli' against different bowling teams
```

```python
run_scored('V Kohli')  # Calling the function to get runs scored by 'V Kohli' against different bowling teams
```

```python
run_scored('DA Warner')
```

```python
mask1=delivery['batsman_runs']==6  # Creating a mask for rows where 'batsman_runs' equals 6
six=delivery[mask1] # Filtering the delivery DataFrame to include only rows where the mask condition is met
```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')  # Creating a pivot table to count the number of 6s scored by each batting team in each over
print(pt)  # Displaying the pivot table
```
